```python
from urllib.parse import quote

# Read the parquet file
df_transform = pd.read_parquet("../files/input/gbif_url_10000.parquet")

# Function to transform URL
def transform_url(original_url):
    """Transform URL to GBIF API image proxy format"""
    # URL encode the original URL
    encoded_url = quote(original_url, safe='')
    # Prepend GBIF API prefix
    return f"https://api.gbif.org/v1/image/unsafe/{encoded_url}"

# Apply transformation to the 'url' column
df_transform['url'] = df_transform['url'].apply(transform_url)

# Display sample of transformed data
display(df_transform.head())

# Save the transformed dataframe
df_transform.to_parquet("../files/input/gbif_url_10000_transformed.parquet")
print(f"Saved transformed dataframe with {len(df_transform)} rows")
```
